```python
import json
with open("work_dirs/ver5_10/segm_ver5_10.json") as f:
    out_data2 = json.load(f)
```

```python

with open("dataset/label_test.json") as f:
    keydata = json.load(f)
```

```python
from mmdet.core import post_processing as po
import numpy as np
import pycocotools.mask as mask_util
import torch
```

```python
iiidx=0
for seg_key in keydata['images']:
    key_idx = seg_key['id']
    inputs = np.zeros(shape=(1, 1024, 1280))
    scores = np.zeros(shape=(1))
    idx =[0]
    i=0
    for seg_data in out_data2:
        if key_idx==seg_data['image_id']:
            if np.sum(mask_util.decode(seg_data['segmentation']))!=0:
                if np.max(inputs) == 0:
                    inputs[0] = mask_util.decode(seg_data['segmentation'])
                    scores[0] = seg_data['score']
                    idx[0] = i
                else:
                    inputs = np.concatenate((inputs,np.expand_dims( mask_util.decode(seg_data['segmentation']), axis=0)),axis=0)
                    scores = np.append(scores,seg_data['score'])
                    idx.append(i)
                i+=1
    cate_labels = np.ones(shape=(inputs.shape[0]))
    sco,lbl,msk,_ = po.mask_matrix_nms(masks=torch.Tensor(inputs).cuda(),labels=torch.Tensor(cate_labels).cuda(),scores=torch.Tensor(scores).cuda(),filter_thr=0.3)
    
    cate_scores_update = sco.cpu().numpy()
    sco_idx=0
    for i_datas in idx:
        if cate_scores_update[sco_idx]<0.25:
            out_data2.pop(i_datas)
        else:
            out_data2[i_datas]['score']=float(cate_scores_update[sco_idx])
        sco_idx+=1
    iiidx+=1
    
    if iiidx%5==0:
        print(sco)
        print(iiidx)
    
```

```python
with open('segm_ver5_10_matnms_scthr025_ver2.json', 'w') as f:
    json.dump(out_data2, f)
```
